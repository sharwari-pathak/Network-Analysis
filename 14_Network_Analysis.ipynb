```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
```

```python
path = "/content/drive/MyDrive/Python_spring24/Dataset/ga_edgelist.csv"
```

```python
df = pd.read_csv(path)
```

```python
df.head()
```

```python
graph = nx.from_pandas_edgelist(df,source = "from",target = 'to')
```

```python
pos = nx.circular_layout(graph)
```

```python
nx.draw(graph,pos)
plt.show()
```

```python
nx.draw(graph,pos,with_labels = True,node_color = 'red',node_size = 1000)
plt.show()
```

```python
pos = nx.random_layout(graph)
```

```python
nx.draw(graph,pos,with_labels = True,node_color = 'red',node_size = 1000)
plt.show()
```

```python
graph = nx.from_pandas_edgelist(df,source = "from",target = 'to',create_using=nx.DiGraph)
```

```python
pos = nx.spring_layout(graph)
```

```python
nx.draw(graph,pos,with_labels = True,node_color = 'red',node_size = 1000)
plt.show()
```

```python
degree = graph.degree()
deg = pd.DataFrame(degree,columns=['Nodes','Degree'])
deg.head(30)
```

```python
import networkx as nx

# Create a graph from the dataset
G = nx.from_pandas_edgelist(df, source='from', target='to')

# Calculate modularity using community detection (Greedy Modularity method)
from networkx.algorithms.community import greedy_modularity_communities
communities = greedy_modularity_communities(G)

# Convert the community structure into a partition for modularity score calculation
modularity_score = nx.algorithms.community.quality.modularity(G, communities)

modularity_score
```
